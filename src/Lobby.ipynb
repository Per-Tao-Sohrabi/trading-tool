```python
from RecursiveScraper import RecursiveScraper
import firebase_admin
from firebase_admin import credentials, firestore

# Initiate a scraper instance
riktskurs_scraper = RecursiveScraper()

# Connect to database
cred = credentials.Certificate("/Users/pertao/playground/trading-tool/api-keys/web-scraper-demo-dd273-firebase-adminsdk-fbsvc-4520a74fff.json")
firebase_admin.initialize_app(cred)

# Get Firestore database instance
db = firestore.client()
```

```python
# Set site to scrap
site_name = "riktkurser.se"

site_data = db.collection("sites").document(site_name).get().to_dict()

# Extract scraping parameters
source_url = site_data["source_url"]
base_url = site_data["base_url"]
key_words = site_data["key_words"]

```

```python
result = riktskurs_scraper.scrap(key_words_dict=key_words,source_url = source_url, base_url=base_url)
print(len(result))
```
